```python
import pandas as pd
import matplotlib.pyplot as plt

# Our video data
data = {
    'Position': [0, 1, 2, 3, 4, 5],
    'Video_B': [103.63, 87.91, 84.54, 81.15, 81.26, 80.00],
    'Video_C': [103.97, 89.17, 85.68, 84.62, 83.14, 81.98],
    'Video_D': [107.83, 87.97, 84.34, 83.31, 82.03, 81.06],
    'Video_E': [103.73, 88.70, 84.93, 82.60, 81.12, 81.20],
    'Video_F': [102.97, 88.64, 84.62, 81.96, 81.47, 79.84]
}

# Create DataFrame and calculate statistics
df = pd.DataFrame(data)
df['Average'] = df[['Video_B', 'Video_C', 'Video_D', 'Video_E', 'Video_F']].mean(axis=1)
df['Std_Dev'] = df[['Video_B', 'Video_C', 'Video_D', 'Video_E', 'Video_F']].std(axis=1)

# Create plot
plt.figure(figsize=(12, 6))

# Plot individual videos
plt.plot(df['Position'], df['Video_B'], 'o-', label='Video B', alpha=0.7)
plt.plot(df['Position'], df['Video_C'], 'o-', label='Video C', alpha=0.7)
plt.plot(df['Position'], df['Video_D'], 'o-', label='Video D', alpha=0.7)
plt.plot(df['Position'], df['Video_E'], 'o-', label='Video E', alpha=0.7)
plt.plot(df['Position'], df['Video_F'], 'o-', label='Video F', alpha=0.7)

# Plot average with error bars
plt.errorbar(df['Position'], df['Average'], yerr=df['Std_Dev'], 
            color='black', linewidth=2, label='Average ± Std Dev',
            capsize=5)

plt.title('Video Retention Analysis')
plt.xlabel('Position')
plt.ylabel('Retention (%)')
plt.grid(True)
plt.legend()

# Show and save plot
plt.savefig('statistical_analysis.png')
plt.show()

# Print results
print("\nStatistical Analysis Results:")
print("-" * 50)

print("\nAverage Retention at Each Position:")
for pos in range(6):
    print(f"Position {pos}:")
    print(f"  Average: {df['Average'][pos]:.2f}%")
    print(f"  Std Dev: {df['Std_Dev'][pos]:.2f}%")

print("\nDrops Between Positions:")
for i in range(1, 6):
    drop = df['Average'][i] - df['Average'][i-1]
    print(f"Drop from position {i-1} to {i}: {drop:.2f}%")
```
